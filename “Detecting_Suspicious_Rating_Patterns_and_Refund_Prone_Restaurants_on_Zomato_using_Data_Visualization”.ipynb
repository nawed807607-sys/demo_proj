```python
#Data Cleaning part
df = df[['rate','votes','approx_cost(for two people)','online_order',
         'book_table','location','rest_type','cuisines',
         'listed_in(type)','listed_in(city)']]
```

```python
df.head()
```

```python
#Clean the rate column as it contains values like "4.1/5", "NEW", "-".
```

```python
df['rate'] = df['rate'].astype(str)
df = df[df['rate'] != 'NEW']
df = df[df['rate'] != '-']

df['rate'] = df['rate'].apply(lambda x: x.split('/')[0])
df['rate'] = df['rate'].astype(float)

df['rate'].head()
```

```python
#now Checking types again
```

```python
df.dtypes
```

```python
#crt dat
import numpy as np

dates = [
    '12-01-2024','15-02-2024','18-03-2024','21-04-2024',
    '25-05-2024','10-06-2024','14-07-2024','19-08-2024',
    '23-09-2024','30-10-2024','11-11-2024','05-12-2024'
]

df['order_date'] = np.random.choice(dates, size=len(df))
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')
```

```python
#cnvrtg it to nrml dat(using pandas to_datetime() to perform time-based analysis.)
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')

df.dtypes
```

```python
#Plot 1: Distribution of Ratings (rate)
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['rate'], kde=True)
plt.title("Distribution of Restaurant Ratings")
plt.show()
```

Most restaurants have ratings between 3.5 and 4.5, indicating generally positive customer feedback on Zomato.

```python
#STEP 11 — Univariate Analysis (Categorical)
```

```python
#Plot 2: Online Order Availability
sns.countplot(x='online_order', data=df)
plt.title("Online Order Availability")
plt.show()
```

Majority of restaurants provide online ordering facility, showing the importance of delivery platforms.

```python
#Bivariate Analysis (Rating vs Cost)
```

```python
#Plot 3: Rating vs Approx Cost
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='rate', y='approx_cost(for two people)', data=df)
plt.title("Rating vs Cost for Two People")
plt.show()
```

Higher cost does not always guarantee higher ratings. Some low-cost restaurants also have excellent ratings.

```python
#Votes vs Rating (trust indicator)
sns.scatterplot(x='votes', y='rate', data=df)
plt.title("Votes vs Rating")
plt.show()
```

Restaurants with higher votes and high ratings are more trustworthy. Low votes with very high ratings may indicate suspicious rating patterns.

```python
#Location vs Ratings (problem areas)
plt.figure(figsize=(10,6))
sns.boxplot(x='listed_in(city)', y='rate', data=df)
plt.xticks(rotation=90)
plt.title("Ratings across City Areas")
plt.show()
```

Certain locations show lower median ratings, which may indicate higher chances of customer complaints and refunds.

```python
#Restaurant Type vs Rating
plt.figure(figsize=(10,6))
sns.boxplot(x='listed_in(type)', y='rate', data=df)
plt.xticks(rotation=90)
plt.title("Restaurant Type vs Rating")
plt.show()
```

Some restaurant types consistently receive lower ratings, which can be flagged for quality checks.

```python
#Using the Date (month analysis)
df['month'] = df['order_date'].dt.month


sns.countplot(x='month', data=df)
plt.title("Orders by Month")
plt.show()
```

Order patterns across months can help identify peak complaint periods.
